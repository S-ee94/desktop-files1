```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

df = pd.read_csv("C:\Seema Python CSV\mtcars.csv")
df
```

```python
x = df["wt"]
plt.title('Histogram for the weight values')
plt.xlabel("weight of the cars")
plt.ylabel("Count")
plt.hist(x,bins=30)
plt.show()
```
