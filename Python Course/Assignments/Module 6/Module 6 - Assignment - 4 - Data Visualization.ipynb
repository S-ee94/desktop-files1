```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

df = pd.read_csv("C:\Seema Python CSV\mtcars.csv")
df
```

```python
x = df["cyl"]
y = df["model"]
fig = plt.figure(figsize =(10, 7)) 
plt.pie(x,labels = y)
plt.title("Pie Chart for Cylinder and Model")
plt.show()
```
