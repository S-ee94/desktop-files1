```python
# 1) Using sklearn import the wine dataset
# 2) Split the data into train and test 
# 3) Train the model
# 4) Make Predictions
# 5) Check accuracy using accuracy_score function
```

```python
import pandas as pd
from sklearn.datasets import load_wine
data = load_wine()
```

```python
y = data.target
y = pd.DataFrame(y)
```

```python
y
```

```python
data
```

```python
data.data
```

```python
dt1 = data.data.tolist()
len(dt1)
```

```python
dt1
```

```python
dt2 = data.feature_names
dt2
```

```python
dta = pd.DataFrame()
dta = pd.DataFrame(dt1,columns = dt2)
x = dta
x.head()
```

```python
y.head()
```

```python
# Splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)  #assuming 30% of data in testing set
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# model are applied for above splitted data - Training the model
# Decision tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dt = dtc.fit(x_train,y_train)
```

```python
# predicting
y_pred_dtc = dt.predict(x_test)
y_pred_dtc
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Random forest 
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rf = rfc.fit(x_train,y_train)
```

```python
y_pred_rfc=rf.predict(x_test)
y_pred_rfc
```

```python
# Linear Regression
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
lrc = lr.fit(x_train,y_train)
lrc
```

```python
lr1 = lrc.predict(x_test)
```

```python
import numpy as np
lr1 = np.hstack(lr1)
lr1 = lr1.astype(int)
lr1
```

```python
y_test.head()
```

```python
print('Accuracy score of Decision tree classifier',accuracy_score(y_test,y_pred_dtc))
```

```python
print('Accuracy score of Random Forest classifier',accuracy_score(y_test,y_pred_rfc))
```

```python
print('Accuracy score of Linear Regression',accuracy_score(y_test,lr1))
```
