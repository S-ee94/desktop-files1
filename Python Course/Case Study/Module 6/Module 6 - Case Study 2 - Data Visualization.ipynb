```python
# 1) Load cars data as dataframe using pandas and create a bar plot between number of cylinders and frequency of cars with that 
# many number of cylinders.
# - Set xlabel as Number of cylinders.
# - Set ylabel as Frequency of cars.
#  - Draw a bar plot.2.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\Seema Python CSV\mtcars.csv")
df
```

```python
x = df["cyl"]
y = df["model"]
plt.xlabel("Number of cylinders")
plt.ylabel("Frequency of cars")
plt.bar(x,y)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:\Seema Python CSV\mtcars.csv")
df
x = df.isnull().sum()
```

```python
plt.bar(x,height = 5)
plt.show()
```

```python
# 3) Use the 'mpg' (Miles Per Gallon column) and draw a histogram
# i) Set xlabel: Miles per gallon
# ii) Set ylabel: Frequency
# iii) Set title as Miles Per Gallon Histogram
# iv) Use mpg column to generate a histogram
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:\Seema Python CSV\mtcars.csv")
df
x = df["mpg"]
plt.xlabel("Miles per gallon")
plt.ylabel("Frequency")
plt.title("Miles Per Gallon Histogram")
plt.hist(x,bins=20)
plt.show()
```
